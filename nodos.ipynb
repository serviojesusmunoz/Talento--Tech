```
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
nodos = ['A1', 'B2', 'C3', 'D4']
aristas = [('A', 'B'), ('B', 'A'), ('B', 'C'), ('C', 'D'),('D', 'A')]
#aqui se crea el grafo
G = nx.Graph()
G.add_edges_from(aristas)
# aqui se dibuja el grafo
plt.figure(figsize=(6, 4))
nx.draw(G, with_labels=True, )
plt.title("Grafo")
plt.show()
# aqui se muestra la matriz adyacente
matriz_adyacencia = nx.adjacency_matrix(G).todense()
print("Matriz de Adyacencia:")
print(matriz_adyacencia)
#aqui se obtiene la matriz de incidencia
matriz_incidencia = nx.incidence_matrix(G, oriented=True).todense()
print("Matriz de Incidencia:")
print(matriz_incidencia)
```
