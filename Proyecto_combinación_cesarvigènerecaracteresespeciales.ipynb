```
import string

# Alfabeto extendido personalizado
ALFABETO = string.ascii_letters + string.digits + "!#$%&/()+-_*{}[]<>.,;:=?¡¿"

def cifrado_cesar(texto, desplazamiento, descifrar=False):
    resultado = ""
    for caracter in texto:
        if caracter in ALFABETO:
            idx = ALFABETO.index(caracter)
            nuevo_idx = (idx - desplazamiento) % len(ALFABETO) if descifrar else (idx + desplazamiento) % len(ALFABETO)
            resultado += ALFABETO[nuevo_idx]
        else:
            resultado += caracter
    return resultado

def cifrado_vigenere(texto, clave, descifrar=False):
    resultado = ""
    clave_repetida = (clave * ((len(texto) // len(clave)) + 1))[:len(texto)]
    for i in range(len(texto)):
        if texto[i] in ALFABETO and clave_repetida[i] in ALFABETO:
            idx_texto = ALFABETO.index(texto[i])
            idx_clave = ALFABETO.index(clave_repetida[i])
            nuevo_idx = (idx_texto - idx_clave) % len(ALFABETO) if descifrar else (idx_texto + idx_clave) % len(ALFABETO)
            resultado += ALFABETO[nuevo_idx]
        else:
            resultado += texto[i]
    return resultado

def cifrar_combinado(texto, cesar, clave_vigenere):
    paso1 = cifrado_cesar(texto, cesar)
    paso2 = cifrado_vigenere(paso1, clave_vigenere)
    return paso2

def descifrar_combinado(texto, cesar, clave_vigenere):
    paso1 = cifrado_vigenere(texto, clave_vigenere, descifrar=True)
    paso2 = cifrado_cesar(paso1, cesar, descifrar=True)
    return paso2

def main():
    print("\n *** PROYECTO ALGORITMOS CESAR-VÌGENERE-CARACTERES ESPECIALES \n COMBINADOS PARA EL BOOTCAMP DE TALENTO TECH ***")

    historial = []

    while True:
        print("\n Opciones:")
        print("1. Cifrar mensaje")
        print("2. Descifrar mensaje anterior")
        opcion = input("Selecciona una opción (1 o 2, o escribe 3 para salir): ")

        if opcion == "3":
            print("Programa finalizado.")
            break

        elif opcion == "1":
            mensaje = input("Digite el mensaje que desea cifrar: ")

            while True:
                try:
                    cesar = int(input("Ingresa el desplazamiento para César (debe ser un número entero): "))
                    break
                except ValueError:
                    print("Por favor, escribe un número entero válido para César.")

            while True:
                clave_entrada = input("Ingresa la clave para Vigenère (mínimo 3 caracteres válidos): ")
                clave_filtrada = ''.join([c for c in clave_entrada if c in ALFABETO])

                if len(clave_filtrada) < 3:
                    print("La clave debe tener al menos 3 caracteres válidos del alfabeto definido.")
                    continue
                else:
                    clave_vigenere = clave_filtrada
                    break

            resultado = cifrar_combinado(mensaje, cesar, clave_vigenere)

            historial.insert(0, {
                "texto": resultado,
                "cesar": cesar,
                "clave": clave_vigenere
            })
            historial = historial[:3]

            print("\n El mensaje a sido cifrado:")
            print(resultado)

            try:
                with open("mensaje_cifrado.txt", "w", encoding="utf-8") as f:
                    f.write(resultado)
            except Exception as e:
                print("Error al guardar el mensaje cifrado:", e)

        elif opcion == "2":
            if not historial:
                print("No existen mensajes para decifrar.")
                continue

            print("\n Mensajes cifrados recientes:")
            for i, item in enumerate(historial):
                print(f"{i + 1}. {item['texto'][:50]}...")

            while True:
                try:
                    eleccion = int(input("Selecciona el número del mensaje a descifrar: "))
                    if 1 <= eleccion <= len(historial):
                        break
                    else:
                        print("Número fuera de rango.")
                except ValueError:
                    print("Ingresa un número válido.")

            seleccionado = historial[eleccion - 1]
            resultado = descifrar_combinado(
                seleccionado["texto"],
                seleccionado["cesar"],
                seleccionado["clave"]
            )

            print("\n Mensaje descifrado:")
            print(resultado)

            try:
                with open("mensaje_descifrado.txt", "w", encoding="utf-8") as f:
                    f.write(resultado)
            except Exception as e:
                print("Error al guardar el mensaje descifrado:", e)

        else:
            print("Opción inválida.")

if __name__ == "__main__":
    main()
```
